```python
class Node:
    def __init__(self,data):   #Traverse in linked list
        self.data = data
        self.next = None
def printLinkedlist(head):
    curr = head
    while(curr!= None):
        print(curr.data,end=" ")
        curr = curr.next
a = Node(5)
b = Node(3)
c = Node(7)
a.next =b
b.next =c
head = a
    
printLinkedlist(head)
    
```

```python
NewNode = Node(6)   #Insertion in linked list at beginning
NewNode.next = head
head = NewNode
printLinkedlist(head)
```

```python
NewNode = Node(9) #Insertion in Linked list at end
curr = head
while(curr.next != None):
    curr = curr.next
curr.next = NewNode
printLinkedlist(head)

```

```python
k = 2            #Insertion in Linked list at kth Index
NewNode = Node(8)
curr = head
for i in range(k-1):
    curr = curr.next
NewNode.next = curr.next
curr.next = NewNode
printLinkedlist(head)

    
```

```python
head = head.next     #Deletion in linked list at beginning
printLinkedlist(head)
```

```python
curr = head                         #Deletion in Linked list at end
while(curr.next.next != None):
    curr = curr.next
curr.next = None
printLinkedlist(head)
```

```python
k =2
curr = head                #Deletion in Linked list at kth Index
for i in range(k-1):
    curr = curr.next
curr.next = curr.next.next
printLinkedlist(head)
```
